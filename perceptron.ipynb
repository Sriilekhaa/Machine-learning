```python
#Multi-layer perceptron with multiple output for Linear and Non-linear datasets and with bias and without bias

import matplotlib.pyplot as plt
import numpy as np
# Activation functions
def sigmoid(x):
  return 1 / (1 + np.exp(-x))

def tanh(x):
  return np.tanh(x)

def relu(x):
  return np.maximum(0, x)

# Derivative of activation functions (for backpropagation)
def sigmoid_derivative(x):
  return sigmoid(x) * (1 - sigmoid(x))

def tanh_derivative(x):
  return 1 - np.tanh(x)**2

def relu_derivative(x):
  return np.where(x > 0, 1, 0)

# Multi-layer perceptron
class MultiLayerPerceptron:
  def __init__(self, input_size, hidden_sizes, output_size, activation='sigmoid', bias=True):
    self.weights = []
    self.biases = []
    self.activation = activation

    layer_sizes = [input_size] + hidden_sizes + [output_size]
    for i in range(1, len(layer_sizes)):
      self.weights.append(np.random.rand(layer_sizes[i-1], layer_sizes[i]))
      if bias:
        self.biases.append(np.random.rand(layer_sizes[i]))
      else:
        self.biases.append(np.zeros(layer_sizes[i]))

  def forward(self, X):
    activations = [X]
    for i in range(len(self.weights)):
      z = np.dot(activations[-1], self.weights[i]) + self.biases[i]
      if self.activation == 'sigmoid':
        a = sigmoid(z)
      elif self.activation == 'tanh':
        a = tanh(z)
      elif self.activation == 'relu':
        a = relu(z)
      else:  # Linear activation
        a = z
      activations.append(a)
    return activations[-1]

# Generate linear dataset
X_linear = np.linspace(-5, 5, 100)
y_linear = 2 * X_linear + 1  # Example linear relationship
y_linear_multi = np.vstack((y_linear, y_linear / 2)).T  # Example multi-output

# Generate non-linear dataset (XOR)
X_nonlinear = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_nonlinear = np.array([0, 1, 1, 0])
y_nonlinear_multi = np.vstack((y_nonlinear, 1 - y_nonlinear)).T  # Example multi-output

# Multi-Layer Perceptron for Linear Dataset (Multiple Outputs)
print("\nMulti-Layer Perceptron for Linear Dataset (Multiple Outputs):")
for bias in [True, False]:
  for activation in ['linear', 'sigmoid', 'tanh', 'relu']:
    mlp = MultiLayerPerceptron(input_size=1, hidden_sizes=[4, 4], output_size=2, activation=activation, bias=bias)
    output = mlp.forward(X_linear.reshape(-1, 1))
    plt.figure()
    for i in range(output.shape[1]):
      plt.plot(X_linear, y_linear_multi[:, i], label=f'Target {i+1}')
      plt.plot(X_linear, output[:, i], label=f'MLP Output {i+1}')
    plt.title(f"Activation: {activation}, Bias: {bias}")
    plt.legend()
    plt.show()

# Multi-Layer Perceptron for Non-linear Dataset (Multiple Outputs)
print("\nMulti-Layer Perceptron for Non-linear Dataset (Multiple Outputs):")
for bias in [True, False]:
  for activation in ['linear', 'sigmoid', 'tanh', 'relu']:
    mlp = MultiLayerPerceptron(input_size=2, hidden_sizes=[4, 4], output_size=2, activation=activation, bias=bias)
    output = mlp.forward(X_nonlinear)
    print(f"Activation: {activation}, Bias: {bias}, Output: {output}")
```

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Preprocess the data
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Define the MLP model
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')  # 10 classes for digits 0-9
])

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
history = model.fit(x_train, y_train, epochs=10, batch_size=32, validation_split=0.2)

# Evaluate the model
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print(f'Test Loss: {test_loss:.4f}')
print(f'Test Accuracy: {test_accuracy:.4f}')

# Plot training & validation accuracy values
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```
